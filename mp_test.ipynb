```python
import sys
import mediapipe as mp
import cv2 as cv2

print("python exe:", sys.executable)
print("mediapipe file:", mp.__file__)
print("mediapipe version:", getattr(mp, "__version__", "unknown"))
print("has solutions:", hasattr(mp, "solutions"))
```

```python
mp_drawing = mp.solutions.drawing_utils #to draw detections from holistic model
mp_holistic = mp.solutions.holistic  # importing the holistic model
mp_drawing_styles = mp.solutions.drawing_styles
```

Get realtime webcam feed

```python
cap = cv2.VideoCapture(0) #0 is the device id that captures the video i.e. webcam
while cap.isOpened():
    ret, frame = cap.read()# reading frame from the video
    cv2.imshow('Raw webcam feed', frame) #showing the frame
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break   # if pressed q gets quit
    
cap.release()
cv2.destroyAllWindows()
```

2. Make Detections from feed 
1. Detect Facial Landmarks
2. Detect Hand Poses
3. Detect Body Poses

```python
cap = cv2.VideoCapture(0) #0 is the device id that captures the video i.e. webcam
with mp_holistic.Holistic(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as holistic:
    while cap.isOpened():
        ret, frame = cap.read()# reading frame from the video
        
        #Recolor feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #Make detections
        results = holistic.process(image)
        #print(results.face_landmarks)
        # face_landmarks, pose_landmarks, left_hand_landmarks, right_hand_landmarks
        
        # Recolor the image back to BGR for rendering
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        #Draw face landmarks
        #mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACE_CONNECTIONS)
        mp_drawing.draw_landmarks(
        image,
        results.face_landmarks,
        mp_holistic.FACEMESH_TESSELATION,
        landmark_drawing_spec=None,
        connection_drawing_spec=mp_drawing_styles
        .get_default_face_mesh_tesselation_style())
        
        
        #right_hand_landmarks
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        
        #left_hand_landmarks
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
        
        #pose_landmarks
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)


        
        
        cv2.imshow('Raw webcam feed', image) #showing the frame
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break   # if pressed q gets quit
    
cap.release()
cv2.destroyAllWindows()
```
